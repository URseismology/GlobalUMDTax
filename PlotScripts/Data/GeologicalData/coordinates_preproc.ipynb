### Method 1, Read Text File and Not XML

```python
import pandas as pd
```

```python
region = 'Proterozoic_Province_2'
df  =  pd.read_csv("coordinates_read_raw.txt", sep=" ", header=None)
df = pd.melt(df)
df = df[['value']]
df['lon'] = df.value.apply(lambda x: x.split(',')[0])
df['lat'] = df.value.apply(lambda x: x.split(',')[1])
df['region'] = region
df[['region','lon','lat']].to_csv('coordinates_read_processed.csv',index=None, header=None)
```

### Method 2, Read Text XML Directly and Parse

```python
import xml.etree.ElementTree as ET
namespaces = {'kml': 'http://www.opengis.net/kml/2.2'}
xpath = ".//kml:Placemark/kml:Polygon/kml:outerBoundaryIs/kml:LinearRing/kml:coordinates"
```

```python
root = ET.parse("/scratch/tolugboj_lab/Sayan_Swar_WS/PythonEnv/Python_Notebooks/GoGlobal/scripts/station_metadata/geological_data/coordinates_read_raw.xml")
region = 'Coastline'

coords=[]
for elm in root.findall(xpath,namespaces):
    coords.append(elm.text.strip())

coords = coords[0].split(' ')
df = pd.DataFrame(coords)
df.rename(columns={0:'value'},inplace=True)
df['lon'] = df.value.apply(lambda x: x.split(',')[0])
df['lat'] = df.value.apply(lambda x: x.split(',')[1])
df['region'] = region
df[['region','lon','lat']].to_csv('/scratch/tolugboj_lab/Sayan_Swar_WS/PythonEnv/Python_Notebooks/GoGlobal/scripts/station_metadata/geological_data/coordinates_read_processed.csv',index=None, header=None)
```
